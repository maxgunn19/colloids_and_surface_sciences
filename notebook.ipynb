## **Problem 1: Two Spherical Bubbles**
---
**Question:** Two spherical bubbles come into contact and fuse in a liquid of hydrostatic pressure P. The radius of the larger bubble is $R_a$ and that of the smaller bubble is $R_b$. There is a curved septum where the two bubbles are in contact. The radius of curvature of the septum is $R_s$. 

**a)** Derive an equation that relates $R_s$ to $R_a$ and $R_b$. (Note: the septum is a thin liquid film, so it has a vapor/liquid interface on both sides.)

**b)** Why is the septum shaped the way it is?

### **1a) Derivation of the Septum Radius Equation**

**Theoretical Framework:**
The pressure difference across a curved interface is described by the **Young-Laplace Equation**. This principle governs the internal pressure of the bubbles and the pressure difference across the septum separating them.

**Young-Laplace Equation for a General Surface:**
$$ \Delta P = \gamma \left( \frac{1}{R_1} + \frac{1}{R_2} \right) $$
Where $\Delta P$ is the pressure difference (Laplace pressure), $\gamma$ is the surface tension, and $R_1$ and $R_2$ are the principal radii of curvature.

**Step 1: Define the Pressure Inside Each Bubble**
For a perfect sphere, the two principal radii of curvature are equal ($R_1 = R_2 = R$). The Young-Laplace equation simplifies to:
$$ \Delta P = \frac{2\gamma}{R} $$
We can use this to define the internal pressure ($P_{internal}$) of each bubble relative to the external liquid pressure ($P$).

For the **larger bubble (a)**, the internal pressure $P_a$ is:
$$ P_a - P = \frac{2\gamma}{R_a} \implies P_a = P + \frac{2\gamma}{R_a} \quad (1) $$

For the **smaller bubble (b)**, the internal pressure $P_b$ is:
$$ P_b - P = \frac{2\gamma}{R_b} \implies P_b = P + \frac{2\gamma}{R_b} \quad (2) $$

**Step 2: Define the Pressure Across the Septum**
The septum is a thin liquid film, meaning it has two vapor/liquid interfaces (one facing bubble 'a' and one facing bubble 'b'). Therefore, the total pressure difference it can sustain is twice that of a single interface. The radius of curvature of this septum is $R_s$.
$$ \Delta P_{\text{septum}} = \frac{4\gamma}{R_s} \quad (3) $$

**Step 3: Equate the Pressure Differences**
The pressure difference across the septum must be equal to the difference between the internal pressures of the two bubbles. Since $R_b < R_a$, the pressure in the smaller bubble is greater ($P_b > P_a$). The septum is therefore curved towards the larger bubble.
$$ P_b - P_a = \Delta P_{\text{septum}} = \frac{4\gamma}{R_s} $$

**Step 4: Substitute and Solve for $R_s$**
We can now create another expression for $P_b - P_a$ by substituting Equations (1) and (2):
$$ P_b - P_a = \left( P + \frac{2\gamma}{R_b} \right) - \left( P + \frac{2\gamma}{R_a} \right) $$
$$ P_b - P_a = \frac{2\gamma}{R_b} - \frac{2\gamma}{R_a} $$
Now, we set our two expressions for the pressure difference equal to each other:
$$ \frac{4\gamma}{R_s} = \frac{2\gamma}{R_b} - \frac{2\gamma}{R_a} $$
Dividing the entire equation by $2\gamma$ simplifies the relationship:
$$ \frac{2}{R_s} = \frac{1}{R_b} - \frac{1}{R_a} $$
Finally, we solve for $R_s$ to get the desired relationship:
$$ \frac{2}{R_s} = \frac{R_a - R_b}{R_a R_b} \implies R_s = \frac{2R_a R_b}{R_a - R_b} $$

**Final Derived Equation:**
$$ R_s = \frac{2R_a R_b}{R_a - R_b} $$

### **1b) Explanation of the Septum's Shape**

The shape of the septum is a direct consequence of the pressure difference between the two bubbles, as dictated by the Young-Laplace equation.

* **Pressure-Radius Relationship:** The Laplace pressure ($\Delta P$) is inversely proportional to the bubble's radius ($R$).
    $$ \Delta P \propto \frac{1}{R} $$
* **Higher Pressure in Smaller Bubble:** Because the smaller bubble has a smaller radius ($R_b < R_a$), it must have a higher internal pressure to maintain its curvature against surface tension ($P_b > P_a$).
* **Resulting Force and Shape:** When the two bubbles fuse, this pressure imbalance creates a net force. The higher pressure from the smaller bubble pushes the dividing wall (the septum) into the region of lower pressure (the larger bubble). This causes the septum to be curved, **bulging into the larger bubble**. Its center of curvature is located inside the smaller bubble, and its precise radius of curvature ($R_s$) is determined by the final equation derived above.

## **Problem 2: Vapor Bubble Nucleation**
---
**Question:** 
**a)** Show that the **Thompson Equation** for nucleation of vapor bubbles in a boiling liquid is:
$$ \Delta H_{vap} \left[ \frac{1}{T_0} - \frac{1}{T} \right] = R \ln \left[ \frac{\frac{2\gamma}{r} + P_l}{P_l} \right] $$
where $P_l$ is the pressure in the bulk liquid phase, set by the ambient pressure of the surrounding gas.

**b)** If the heat of vaporization is **31 kJ/mol** and the surface tension of water is **18.4 mN/m** at **272 °C**, find the radius of a vapor bubble in equilibrium with liquid at an ambient pressure of **1 atm** and T = **272 °C**.

### **2a) Derivation of the Thompson Equation**

**Theoretical Framework:**
This derivation connects thermodynamics (Gibbs-Duhem), phase equilibria (Clausius-Clapeyron), and surface phenomena (Young-Laplace) to describe the conditions required for a vapor bubble of radius `r` to exist in a superheated liquid.

**Step 1: Start with the Gibbs-Duhem Equation**
The Gibbs-Duhem equation relates changes in pressure (P), temperature (T), and chemical potential ($\mu$) for a given phase. We write it for the liquid (l) and vapor/gas (g) phases.
$$ S_l dT - V_l dP_l + d\mu_l = 0 $$
$$ S_g dT - V_g dP_g + d\mu_g = 0 $$

**Step 2: Apply Equilibrium Condition**
At equilibrium, the chemical potentials of the two phases are equal ($d\mu_l = d\mu_g$). We can therefore set the two equations equal to each other:
$$ S_l dT - V_l dP_l = S_g dT - V_g dP_g $$
Rearranging the terms gives:
$$ (S_g - S_l)dT = V_g dP_g - V_l dP_l \quad (1) $$

**Step 3: Introduce Key Physical Relationships**
To proceed, we introduce three essential equations:
1.  **Clausius Equation:** Relates the entropy of vaporization ($\Delta S_{vap} = S_g - S_l$) to the enthalpy of vaporization ($\Delta H_{vap}$).
    $$ S_g - S_l = \frac{\Delta H_{vap}}{T} \quad (2) $$
2.  **Young-Laplace Equation:** Relates the pressure inside the vapor bubble ($P_g$) to the pressure in the surrounding liquid ($P_l$).
    $$ P_g - P_l = \frac{2\gamma}{r} \implies P_g = P_l + \frac{2\gamma}{r} \quad (3) $$
3.  **Ideal Gas Law:** We assume the vapor inside the small bubble behaves as an ideal gas.
    $$ V_g = \frac{RT}{P_g} \quad (4) $$

**Step 4: Simplify and Integrate**
We assume the volume of the liquid ($V_l$) is negligible compared to the vapor ($V_g$), and that the external liquid pressure is constant ($dP_l \approx 0$). Applying these to Equation (1):
$$ (S_g - S_l)dT \approx V_g dP_g $$
Substitute the Clausius Equation (2) and the Ideal Gas Law (4):
$$ \left( \frac{\Delta H_{vap}}{T} \right) dT = \left( \frac{RT}{P_g} \right) dP_g $$
Separate the variables for integration:
$$ \frac{\Delta H_{vap}}{R} \frac{dT}{T^2} = \frac{dP_g}{P_g} $$
We integrate from the normal boiling point ($T_0$, where the vapor pressure equals the external liquid pressure $P_l$) to the superheated temperature ($T$, where the vapor pressure inside the bubble is $P_g$):
$$ \frac{\Delta H_{vap}}{R} \int_{T_0}^{T} \frac{dT}{T^2} = \int_{P_l}^{P_g} \frac{dP}{P} $$
Performing the integration yields the **Clausius-Clapeyron Equation**:
$$ \frac{\Delta H_{vap}}{R} \left[ -\frac{1}{T} \right]_{T_0}^{T} = [\ln(P)]_{P_l}^{P_g} $$
$$ \frac{\Delta H_{vap}}{R} \left( \frac{1}{T_0} - \frac{1}{T} \right) = \ln \left( \frac{P_g}{P_l} \right) \quad (5) $$

**Step 5: Final Substitution**
Finally, substitute the expression for $P_g$ from the Young-Laplace Equation (3) into the Clausius-Clapeyron Equation (5):
$$ \frac{\Delta H_{vap}}{R} \left( \frac{1}{T_0} - \frac{1}{T} \right) = \ln \left( \frac{P_l + \frac{2\gamma}{r}}{P_l} \right) $$
Multiplying by R gives the final **Thompson Equation**:
$$ \Delta H_{vap} \left[ \frac{1}{T_0} - \frac{1}{T} \right] = R \ln \left[ \frac{\frac{2\gamma}{r} + P_l}{P_l} \right] $$

### **2b) Calculation of the Vapor Bubble Radius**

**Equation Rearrangement:**
To solve for the radius `r`, we must first rearrange the Thompson equation. 

1.  Divide by R:
$$ \frac{\Delta H_{vap}}{R} \left[ \frac{1}{T_0} - \frac{1}{T} \right] = \ln \left[ 1 + \frac{2\gamma}{rP_l} \right] $$
2.  Exponentiate both sides to remove the natural logarithm:
$$ \exp\left( \frac{\Delta H_{vap}}{R} \left[ \frac{1}{T_0} - \frac{1}{T} \right] \right) = 1 + \frac{2\gamma}{rP_l} $$
3.  Isolate the term containing `r`:
$$ \frac{2\gamma}{rP_l} = \exp\left( \frac{\Delta H_{vap}}{R} \left[ \frac{1}{T_0} - \frac{1}{T} \right] \right) - 1 $$
4.  Solve for `r`:
$$ r = \frac{2\gamma}{P_l \left( \exp\left( \frac{\Delta H_{vap}}{R} \left[ \frac{1}{T_0} - \frac{1}{T} \right] \right) - 1 \right)} $$

```python
# === Define Constants and Given Data ===
# All values are converted to base SI units for calculation.

# H_vap: Heat of vaporization for water (31 kJ/mol -> J/mol)
H_vap = 31e3

# gamma: Surface tension of water at 272 C (18.4 mN/m -> N/m)
gamma = 18.4e-3

# T: Superheated temperature of the liquid (272 C -> K)
T = 272 + 273.15

# P_l: Ambient pressure of the bulk liquid (1 atm -> Pa)
P_l = 101325

# T0: Normal boiling point of water at 1 atm (100 C -> K)
T0 = 100 + 273.15

# R: Ideal gas constant in J/(mol·K)
R = 8.314

# === Calculation ===
# This function implements the rearranged Thompson equation to find the bubble radius.
def calculate_bubble_radius(H_vap, T0, T, R, gamma, P_l):
    """Calculates the equilibrium vapor bubble radius using the Thompson equation."""
    # Calculate the argument for the exponential function
    exponent_term = (H_vap / R) * (1/T0 - 1/T)
    
    # Calculate the denominator of the main equation
    denominator = P_l * (np.exp(exponent_term) - 1)
    
    # Calculate the final radius 'r'
    r = (2 * gamma) / denominator
    
    return r

# Execute the calculation
radius_m = calculate_bubble_radius(H_vap, T0, T, R, gamma, P_l)

# Convert radius from meters to nanometers for readability
radius_nm = radius_m * 1e9

# === Output the Result ===
print(f"The radius of the vapor bubble is {radius_m:.2e} m, or {radius_nm:.2f} nm.")
```

## **Problem 3: NaCl Particle Supersaturation**
---
**Question:** NaCl particles (density = 2.17 g/cm³) with a specific surface area of 4.25 × 10⁵ cm²/g show a supersaturation of 6.71 % in ethanol (T=25 °C). 

**a)** Assuming the crystals are uniform spheres, find their radius (r).

**b)** Estimate the NaCl/ethanol interfacial tension ($\gamma$) from this solubility data.

### **3a) Calculation of Particle Radius**

**Theoretical Framework:**
The specific surface area ($A_{sp}$) of uniform spheres is geometrically related to their density ($\rho$) and radius ($R_s$).

The total surface area of a single sphere is $4\pi R_s^2$.
The mass of a single sphere is $\text{Volume} \times \text{Density} = \frac{4}{3}\pi R_s^3 \rho$.

$A_{sp}$ is defined as the total surface area per unit mass:
$$ A_{sp} = \frac{\text{Surface Area}}{\text{Mass}} = \frac{4\pi R_s^2}{\frac{4}{3}\pi R_s^3 \rho} = \frac{3}{\rho R_s} $$

**Equation Rearrangement:**
We can rearrange this formula to solve for the radius, $R_s$:
$$ R_s = \frac{3}{A_{sp} \rho} $$

```python
# === Define Constants and Given Data ===
# Units are kept in the cgs system (cm, g, s) as provided in the problem.

# A_sp: Specific surface area of NaCl particles in cm^2/g
A_sp = 4.25e5

# rho_NaCl: Density of NaCl in g/cm^3
rho_NaCl = 2.17

# === Calculation ===
# The radius Rs is calculated using the rearranged formula.
Rs_cm = 3.0 / (A_sp * rho_NaCl)

# Convert the radius from centimeters to nanometers for better readability
Rs_nm = Rs_cm * 1e7  # (1 cm = 1e7 nm)

# === Output the Result ===
print(f"The radius of the NaCl particles is {Rs_cm:.2e} cm, or {Rs_nm:.2f} nm.")
```

### **3b) Estimation of Interfacial Tension**

**Theoretical Framework:**
The increased solubility of small particles compared to a flat surface is described by the **Ostwald-Freundlich Equation** (a form of the Kelvin equation applied to solubility). This equation relates interfacial tension ($\gamma$) to the particle radius ($R_s$) and the degree of supersaturation ($S/S_0$).

For an electrolyte that dissociates into `m` cations and `n` anions (like NaCl, where m=1, n=1), the equation is:
$$ \frac{2M\gamma}{\rho R_s} = (m + n) RT \ln\left(\frac{S}{S_0}\right) $$

**Equation Rearrangement:**
We rearrange the equation to solve for the interfacial tension, $\gamma$:
$$ \gamma = \frac{(m+n)RT\rho R_s \ln\left(\frac{S}{S_0}\right)}{2M} $$

```python
# === Define Constants and Given Data ===
# All values are converted to base SI units (m, kg, s, K, mol).

# For NaCl, which dissociates into Na+ and Cl-:
m = 1.0  # Number of cations
n = 1.0  # Number of anions

# Supersaturation is given as 6.71%
supersaturation = 0.0671
S_S0 = 1.0 + supersaturation  # The ratio of solubility (S) to bulk solubility (S0)

# R: Ideal gas constant in J/(mol·K)
R = 8.314

# T: Temperature in Kelvin (25 C -> K)
T = 25 + 273.15

# rho: Density of NaCl (2.17 g/cm^3 -> kg/m^3)
rho = 2170

# Rs: Radius of particles calculated in part (a), converted from cm to m.
Rs = Rs_cm * 1e-2  # (1 cm = 1e-2 m)

# M: Molar mass of NaCl (58.44 g/mol -> kg/mol)
M = 58.44e-3

# === Calculation ===
# The interfacial tension (gamma) is calculated using the rearranged Ostwald-Freundlich equation.
gamma_Nm = ((m + n) * R * T * rho * Rs * np.log(S_S0)) / (2 * M)

# Convert the result from N/m to mN/m for conventional reporting
gamma_mNm = gamma_Nm * 1e3

# === Output the Result ===
print(f"The estimated NaCl/ethanol interfacial tension is {gamma_Nm:.3f} N/m, or {gamma_mNm:.2f} mN/m.")
```

## **Problem 4: Spreading of Isoamyl Alcohol and Water**
---
**Question:** The surface tension of isoamyl alcohol (liquid "B") and its interfacial tension with water (liquid "A") at T = 25 °C are listed below. $\gamma_B$ is the surface tension of pure B, while $\gamma_{B(A)}$ represents the surface tension of B that is saturated with A, and $\gamma_{A(B)}$ represents the surface tension of A that is saturated with B. $\gamma_{AB}$ represents the A/B interfacial tension. The surface tension of water at 25 °C is 72.2 mN/m.

**Data:**
* $\gamma_A$ (pure water) = 72.2 mN/m
* $\gamma_B$ (pure isoamyl alcohol) = 23.7 mN/m
* $\gamma_{A(B)}$ (water saturated with B) = 25.9 mN/m
* $\gamma_{B(A)}$ (isoamyl alcohol saturated with A) = 23.6 mN/m
* $\gamma_{AB}$ (interfacial) = 5.0 mN/m

**a)** Will isoamyl alcohol spread on water?

**b)** Will water spread on isoamyl alcohol?

**c)** Will either situation produce the “strange” behavior where a liquid initially spreads, only to retract into a lens after time?

### **4a) Will isoamyl alcohol spread on water?**

**Theoretical Framework:**
Spontaneous spreading occurs if the process leads to a decrease in the total surface free energy. This condition is quantified by the **spreading coefficient (S)**. A positive value (S > 0) indicates spontaneous spreading.

For a liquid B (isoamyl alcohol) spreading on a substrate A (water), the equation is:
$$ S_{B/A} = \gamma_A - (\gamma_B + \gamma_{AB}) $$
This can be interpreted as the surface tension of the substrate minus the sum of the surface tensions of the spreading liquid and the new interface that is created.

```python
# === Define Given Data ===
gamma_A = 72.2         # Surface tension of pure water (substrate) in mN/m
gamma_B = 23.7         # Surface tension of pure isoamyl alcohol (spreading liquid) in mN/m
gamma_AB = 5.0         # Interfacial tension between A and B in mN/m

# === Calculation ===
# Calculate the initial spreading coefficient for B on A.
S_BA = gamma_A - (gamma_B + gamma_AB)

# === Output and Conclusion ===
print(f"Initial Spreading Coefficient (S_B/A): {S_BA:.2f} mN/m")
if S_BA > 0:
    print("Since S > 0, isoamyl alcohol will spontaneously spread on the surface of pure water.")
else:
    print("Since S <= 0, isoamyl alcohol will not spread on water; it will form a lens.")
```

### **4b) Will water spread on isoamyl alcohol?**

**Theoretical Framework:**
The same principle applies, but the roles of substrate and spreading liquid are reversed. For a liquid A (water) spreading on a substrate B (isoamyl alcohol), the equation is:
$$ S_{A/B} = \gamma_B - (\gamma_A + \gamma_{AB}) $$

```python
# === Define Given Data ===
# The variables gamma_A, gamma_B, and gamma_AB are already defined from the previous cell.

# === Calculation ===
# Calculate the initial spreading coefficient for A on B.
S_AB = gamma_B - (gamma_A + gamma_AB)

# === Output and Conclusion ===
print(f"Initial Spreading Coefficient (S_A/B): {S_AB:.2f} mN/m")
if S_AB > 0:
    print("Since S > 0, water will spontaneously spread on the surface of pure isoamyl alcohol.")
else:
    print("Since S <= 0, water will not spread on isoamyl alcohol; it will form a lens.")
```

### **4c) Will retraction behavior occur?**

**Theoretical Framework:**
The "strange" behavior of initial spreading followed by retraction into a lens occurs if the **initial spreading coefficient is positive**, but the **final spreading coefficient (after the liquids have mutually saturated each other) is negative**.

From part (a), we know isoamyl alcohol initially spreads on water ($S_{B/A} > 0$). We must now calculate the final spreading coefficient, $S'_{B/A}$, using the surface tension values for the saturated phases.

The final spreading coefficient for B on A is:
$$ S'_{B/A} = \gamma_{A(B)} - (\gamma_{B(A)} + \gamma_{AB}) $$

```python
# === Define Given Data for Saturated State ===
gamma_A_sat_B = 25.9    # Surface tension of water saturated with isoamyl alcohol
gamma_B_sat_A = 23.6    # Surface tension of isoamyl alcohol saturated with water
# gamma_AB is unchanged

# === Calculation ===
# Calculate the final spreading coefficient for B on A using saturated values.
S_final_BA = gamma_A_sat_B - (gamma_B_sat_A + gamma_AB)

# === Output and Conclusion ===
print(f"Initial Spreading Coefficient (S_B/A) = {S_BA:.2f} mN/m")
print(f"Final Spreading Coefficient (S'_B/A) = {S_final_BA:.2f} mN/m")

# Check the conditions for retraction behavior
if S_BA > 0 and S_final_BA < 0:
    print("Yes, this situation will produce the retraction behavior.")
    print("\nReasoning:")
    print("1. Initial State (S > 0): Isoamyl alcohol initially spreads across the pure water surface to lower the system's energy.")
    print("2. Saturation Process: As it spreads, the two liquids begin to dissolve into one another at the interface, changing their respective surface tensions.")
    print("3. Final State (S' < 0): Once the water surface is saturated with isoamyl alcohol (and vice-versa), the spreading condition is no longer favorable. The spread film retracts into a lens to achieve the new, most stable energetic state.")
else:
    print("No, this situation will not produce the retraction behavior.")
```

## **Problem 5: Critical Surface Tension of Teflon**
---
**Question:** If $\gamma_L$ and $\theta$ are measured for a homologous series of liquids on a given low-energy solid surface, a plot of $\cos\theta$ vs. $\gamma_L$ results in a straight line. 

**a)** Estimate the critical surface tension for a Teflon surface with the following wetting characteristics.

**b)** Is this value a good estimate of the surface energy (tension) of the Teflon?

| Compound      | $\gamma_L$ (mN/m) | $\theta$ (deg) |
|---------------|--------------------|-----------------|

| tetradecane   | 26.7               | 44              |
| dodecane      | 25.4               | 42              |
| undecane      | 24.7               | 39              |

### **5a) Estimating Critical Surface Tension with a Zisman Plot**

**Theoretical Framework:**
The **Zisman plot** is an empirical method to estimate the **critical surface tension ($\gamma_c$)** of a low-energy solid. The critical surface tension is defined as the surface tension of a hypothetical liquid that would just completely wet the solid (i.e., have a contact angle of $\theta = 0°$).

The method involves plotting $\cos\theta$ for a series of non-interacting liquids against their known surface tensions, $\gamma_L$. A linear trend is typically observed, which can be extrapolated to the point where $\cos\theta = 1$ (since $\cos(0°) = 1$). The corresponding $\gamma_L$ value at this intercept is the critical surface tension, $\gamma_c$. 

**Methodology:**
1.  Collect data for $\gamma_L$ and $\theta$ for several liquids on the Teflon surface.
2.  Calculate $\cos\theta$ for each angle. Note that for pentane, "spreads" implies $\theta = 0°$, so $\cos\theta = 1$.
3.  Plot $\cos\theta$ (y-axis) versus $\gamma_L$ (x-axis).
4.  Perform a linear regression on the data points to get an equation of the form $\cos\theta = m \cdot \gamma_L + c$.
5.  Extrapolate this line to find the value of $\gamma_L$ when $\cos\theta = 1$. This value is $\gamma_c$.

```python
# === Data Preparation ===

# surface_tensions: Surface tensions of the homologous series of liquids (in mN/m).
surface_tensions = np.array([27.6, 26.7, 25.4, 24.7, 23.9, 22.9, 21.8, 20.3, 18.4]) # Excludes pentane for the fit

# thetas_deg: Corresponding contact angles on Teflon (in degrees).
# Pentane spreads, so its angle is 0, giving cos(theta) = 1. It represents the target, not part of the fit.
thetas_deg = np.array([46, 44, 42, 39, 35, 32, 26, 21, 12])

# Calculate the cosine of the angles. NumPy's trig functions require input in radians.
cos_thetas = np.cos(np.radians(thetas_deg))

# === Linear Regression and Extrapolation Function ===
def generate_zisman_plot(gamma_L_data, cos_theta_data):
    """
    Performs a linear fit on the provided surface tension and contact angle data,
    extrapolates to find the critical surface tension, and generates a Zisman plot.
    """
    # --- Analysis ---
    # Fit a 1st-degree polynomial (a line) to the data. coeffs = [slope, intercept]
    coeffs = np.polyfit(gamma_L_data, cos_theta_data, 1)
    fit_function = np.poly1d(coeffs) # Creates a callable function from the coefficients
    
    # --- Extrapolation ---
    # The critical surface tension (gamma_c) is where the line intersects cos(theta) = 1.
    # 1 = m*gamma_c + c  => gamma_c = (1 - c) / m
    slope, intercept = coeffs
    critical_surface_tension = (1 - intercept) / slope

    # --- Plotting Setup ---
    # Create an extended range for the x-axis to clearly show the extrapolation.
    plot_x_values = np.linspace(critical_surface_tension - 1, max(gamma_L_data) + 1, 100)
    
    plt.figure(figsize=(10, 7))
    
    # Plot the original experimental data points
    plt.scatter(gamma_L_data, cos_theta_data, color='blue', label='Experimental Data', zorder=5)
    
    # Plot the linear fit/extrapolation line
    plt.plot(plot_x_values, fit_function(plot_x_values), color='red', linestyle='-', label='Linear Fit & Extrapolation')
    
    # Add a horizontal reference line at cos(theta) = 1 (complete wetting)
    plt.axhline(y=1, color='green', linestyle=':', label='Complete Wetting (cos(θ) = 1)')
    
    # Add a vertical line marking the calculated critical surface tension
    plt.axvline(x=critical_surface_tension, color='purple', linestyle='--', 
                label=f'$\gamma_c$ = {critical_surface_tension:.2f} mN/m')
    
    # --- Formatting ---
    plt.title('Zisman Plot for Teflon', fontsize=16)
    plt.xlabel('Liquid Surface Tension, $\gamma_L$ (mN/m)', fontsize=12)
    plt.ylabel('Cosine of Contact Angle, cos(θ)', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xlim(left=15)
    plt.ylim(bottom=0.6)

    # Calculate and display the R^2 value on the plot to show goodness of fit
    correlation_matrix = np.corrcoef(gamma_L_data, cos_theta_data)
    r_squared = correlation_matrix[0, 1]**2
    plt.text(22, 0.65, f'$R^2 = {r_squared:.4f}$', fontsize=12, color='black', 
             bbox=dict(facecolor='white', alpha=0.5))
    
    # Display the plot
    plt.show()
    
    return critical_surface_tension

# === Execution ===
# Generate the plot and capture the calculated critical surface tension
gamma_critical = generate_zisman_plot(surface_tensions, cos_thetas)

# === Output the Result ===
print(f'Estimated Critical Surface Tension: {gamma_critical:.2f} mN/m')
```

### **5b) Is this a good estimate?**
