---
**Question:** Two sperical bubbles come into contact and fuse in a liquid of hydrostatic pressure P. The radius of the larger bubble is $R_a$ and that of the smaller bubble is $R_b$. There is a curved septum where the two bubbles are in contact. The radius of curvature of the septum is $R_s$.  
**a)** Derive an equation that relates $R_s$ to $R_a$ and $R_b$. (Note: the septum is a thin liquid film, so it has a vapor/liquid interface on both sides.)  
**b)** Why is the septum shaped the way that it is?

### **1a) Derivation of the Septum Radius Equation**  ###  
  
  **Theoretical Framework:** The pressure difference across a cruved interface is described by the **Young-Laplace Equation**. This principle governs the internal pressure of each bubble and their difference across the septum which separates them.
  
**Young-Laplace Equation for a General Surface**
$$
\Delta P = \gamma(\frac{1}{R_1} + \frac{1}{R_2})
$$  
  
Where $\Delta P$ is the pressure difference (Laplace pressure), $\gamma$ is the surface tension, and $R_1$ and $R_2$ are the principal radii of curvature.  
  
**Step 1: Define the Pressure Inside Each Bubble** For a perfect sphere, both radii of curvature are equal to one another ($R_1 = R_2 = R$). The Young-Laplace equation simplifies to:

$$
\Delta P = \gamma (\frac{1}{R} + \frac{1}{R}) = \frac{2 \gamma}{R}
$$
  
We can use this to define the internal pressure ($P_{internal}$) of each bubble relative to the external liquid pressure (P).

For the **larger bubble (a)**, the internal pressure $P_a$ is:

$$
P_a - P = \frac{2 \gamma}{R_a} \implies P_a = P + \frac{2 \gamma}{R_a} \tag{1}
$$

For the **smaller bubble (b)**, the internal pressure $P_b$ is:

$$
P_b - P = \frac{2 \gamma}{R_b} \implies P_b = P + \frac{2 \gamma}{R_b} \tag{2}
$$

**Step 2: Define the Pressure Across the Septum** The septum is the film that separates the two bubbles. This film has a liquid/vapor interface on both sides (one facing bubble 'a' and one facing bubble 'b'). Therefore, the pressure difference it can sustain is twice that of a single interface:

$$
\Delta P_{septum} = \frac{4 \gamma}{R_s} \tag{3}
$$

**Step 3: Equate the Pressure Difference** The pressure difference must be equal to the difference between the internal pressures of the two bubbles. Since $R_b < R_a$, the pressure in the smaller bubble is greater ($P_b > P_a$). The septum is therefore curved towards the larger bubble. 

$$
P_b - P_a = \Delta P_{septum} = \frac{4 \gamma}{R_s}
$$

**Step 4: Substitute and Solve for $R_s$** We can now find another expression for the pressure difference $P_b - P_a$ using **Equation (1)** and **Equation (2)**:

$$
P_b - P_a = (P + \frac{2 \gamma}{R_b}) - (P + \frac{2 \gamma}{R_a})
$$

Simplifying we get:

$$
P_b - P_a = \frac{2\gamma}{R_b}-\frac{2\gamma}{R_a}
$$

We can now plug **Equation 3** in:

$$
\frac{4\gamma}{R_s} = \frac{2\gamma}{R_b} - \frac{2\gamma}{R_a}
$$

Dividing the entire equation by by $2\gamma$ simplifies the relationship:

$$
\frac{2}{R_s} = \frac{1}{R_b} - \frac{1}{R_a}
$$

Finally, we can solve for **R_s** in relation to **R_a** and **R_b**.  
  
$$
\frac{2}{R_s} = \frac{R_a - R_b}{R_a R_b} \implies R_s = \frac{2 R_a R_b}{R_a - R_b}
$$

**Final Derived Equation:**

$$
R_s = \frac{2R_aR_b}{R_a - R_b}
$$

### **1b) Explanation of the Septum's Shape**  ###
The shape of the septum is due to the pressure difference between the two bubbles, as dictated by the Young-Laplace equation.  
  
* **Pressure and Radius Relationship:** The Laplace Equation shows that the Laplace pressure ($\Delta P$) inside a bubble is **inversly proportional** to its radius 

$$
\Delta P \propto 1/R
$$
* **Higher Pressure in Smaller Bubble:** Because the smaller bubble has a smaller radius ($R_b < R_a$), it has a **higher internal pressure** ($P_b > P_a).
* **Resulting Force:** When the bubbles touch, the pressure difference creates a net force on the wall between them. The higher pressure of the smaller bubble pushes the septum into the larger bubble.  
  
The septum is curved **bulging into the larger bubble**, with its center of curvature located inside the smaller bubble. The exact radius of the curvature, $R_s$, is determined by the balance between the surface tension forces and the pressure difference.

**a)** Show that the **Thompson Equation** for nucleation of vapor bubbles in a boiling liquid is:

$$  
where $P_l$ is the pressure in the bulk liquid phase, set by the ambient pressure of the surrounding gas.  
**b)** If the heat of vaporization is **31 kJ/mol** and the surface tension of the water is **18.4 mN/m** at **272 $\degree C$**, find the radius of the vapor bubble in equilibrium with liquid at an ambient pressure of **1 atm** and **T = 272 $\degree C$**.

### **2a) Derivation of the Thompson Equation**  ###
  
**Theoretical Framework:** This derivation connected thermodynamics (Gibbs-Duhem), phase equilibria (Clausius-Clapeyron), and the surface phenomena (Young-Laplace) to describe the conditions required for a vapor bubble of radius **r** to exist in a superheatd liquid.  
  
**Step 1: Start with the Gibbs-Duhem Equation** The Gibbs-Duhem equation relates change in pressure (P), temperature (T), and chemical potential ($mu$) for a given phase. We write it for the liquid (l) and vapor/gas (g) phases. will 

$$
\bar{S_l} dT - \bar{V_l} dP_l + d\mu_l = 0
$$

$$
\bar{S_g} dT - \bar{V_g} dP_g + d\mu_g = 0
$$

**Step 2: Combine** They both equal 0 so they can be set equal to one another:

$$
\bar{S_l} dT - \bar{V_l} dP_l + d\mu_l = \bar{S_g} dT - \bar{V_g} dP_g + d\mu_g
$$

Now take the difference:

$$
[\bar{S_l} - \bar{S_g}]dT - \bar{V_l}dP_l + \bar{V_g}dP_g + d[\mu_l - \mu_g] = 0
$$

**Step 3: Apply Equilibrium Condition** At equilibrium, the chemical potential difference is zero ($[d\mu_l - \mu_g] = 0$). That leaves us with the following: 

$$
[\bar{S_l} - \bar{S_g}]dT - \bar{V_l}dP_l + \bar{V_g}dP_g  = 0 
$$
  
Rearranging the terms gives:

$$
[\bar{S_l} - \bar{S_g}]dT = \bar{V_l}dP_l - \bar{V_g}dP_g \tag{1}
$$
  
**Step 4: Introducing Key Physical Relationships** To proceed, we introduce three essential equations:

1. **Clausius Equation**: Relates entropy change of vaporization ($\Delta S_{vap} = S_g - S_l$) to the enthalpy of vaporization ($\Delta H_{vap}$).

$$
\bar{S_g} - \bar{S_l} = \frac{\Delta H_{vap}}{T} \tag{2}
$$

2. **Laplace Pressure Equation**: Describes the pressure inside the vapor bubble ($P_g$) to the pressure in the surrounding liquid ($P_l$).

$$
P_g - P_l = \frac{2\delta}{r} \implies P_g = P_l + \frac{2\gamma}{r} \tag{3}

$$

3. **Ideal Gas Law**: We assume the vapor in the bubble behaves as an ideal gas.

$$
\bar{V}_g = \frac{RT}{P_g} \tag{4}
$$

**Step 4: Simplify and Integrate** We assume the volume of the liquid ($V_l$) is negligible ocmpared to the vapor ($V_g$), and that the external liquid pressure is constant ($dP_l \approx 0$). Applying these to **Equation (1)**:

$$
[\bar{S_l} - \bar{S_g}]dT = - \bar{V_g}dP_g 
$$

Substitute the **Clausius Equation (2)** and the **Ideal Gas Law (4)**

$$
(-\frac{\Delta H_{vap}}{T})dT = - \bar{V}_g dP_g
$$

Simplifying we get:

$$
(\frac{\Delta H_{vap}}{T})dT = \bar{V}_g dP_g
$$

We can assume the vapor in the bubble behaves as an idea gas. That means that we can substitute in the **Ideal Gas Law (4)**:

$$
\frac{\Delta H_{vap}}{T} dT = (\frac{RT}{P_g})dP_g
$$

Separate the variables for integration:

$$
\frac{\Delta H_{vap}}{R}\frac{dT}{T^2} = \frac{dP_g}{P_g}
$$

We integrate from the normal boiling point ($T_0$) to the final superheated temperature (T) and from the normal boiling point (which is the external liquid pressure, $P_l$) to the pressure inside the bubble ($P_g$).

$$
\frac{\Delta H_{vap}}{R} \int_{T_0}^{T} \frac{dT}{T^2} = \int_{P_l}^{P_g} \frac{dP}{P}
$$

Performing the integration gives us the **Clausius-Clapeyron Equation**:  

$$
\frac{\Delta H_{vap}}{R}[-\frac{1}{T}]_{T_0}^{T} = [\ln(P)]_{P_l}^{P_g}
$$


$$
\frac{\Delta H_{vap}}{R}(-\frac{1}{T}-(-\frac{1}{T_0})) = \ln (P_g) - \ln (P_l)
$$


$$
\frac{\Delta H_{vap}}{R}(\frac{1}{T_0} - \frac{1}{T}) = \ln(\frac{P_g}{P_l})
$$

**Step 5: Final Substitution** Finally, we can now substitute in the **Laplace Pressure Equation (3)**:

$$
\frac{\Delta H_{vap}}{R}(\frac{1}{T_0} - \frac{1}{T}) = \ln(\frac{P_l + \frac{2\gamma}{r}}{P_l})
$$

Moving the R over to the right side we are now left with the **Thompson Equation for nucleation of a vapor bubble**:

$$
\Delta H_{vap} [\frac{1}{T_0}-\frac{1}{T}] = R \ln[\frac{\frac{2\gamma}{r} + P_l}{P_l}]
$$  


**b)** If the heat of vaporization is **31 kJ/mol** and the surface tension of the water is **18.4 mN/m** at **272** $\degree C$, find the **radius of the vapor bubble** in equilibrium with liquid at an ambient pressure of **1 atm** and **T = 272** $\degree C$.

### **2b) Calculation of the Vapor Bubble Radius** ###

We will start with the **Thompson Equation for nucleation of a vapor bubble**:

$$
\Delta H_{vap} [\frac{1}{T_0}-\frac{1}{T}] = R \ln[\frac{\frac{2\gamma}{r} + P_l}{P_l}]
$$

**<center>We need to rearrange this equation to solve for r</center>**

1. Simplify the **natural log (ln)** portion of the equation:  

$$
\frac{\frac{2\gamma}{r} + P_l}{P_l} \implies \frac{2\gamma}{rP_l} + \frac{P_l}{P_l} \implies 1 + \frac{2\gamma}{rP_l}
$$  
  
<center> Isolate natural log (ln) </center>

$$
\frac{\Delta H_{vap}}{R}[\frac{1}{T_0} - \frac{1}{T}] = ln[1 + \frac{2\gamma}{rP_l}]
$$

2. Eliminate the **natural log (ln)**:

$$
\exp(\frac{\Delta H_{vap}}{R}[\frac{1}{T_0}-\frac{1}{T}]) = 1 + \frac{2\gamma}{rP_l}
$$

3. Isolate and solve for r:

$$
r = \frac{2\gamma}{P_l(\exp(\frac{\Delta H_{vap}}{R}[\frac{1}{T_0}-\frac{1}{T}])-1)}
$$

```python
# === Defining Constants and Variables ===
# Constants and Data
H_vap = 31e3        # H_vap: Heat of vaporization in J/mol (31 kJ/mol -> J/mol)
gamma = 18.4e-3     # gamma: Surface tension of water at 272 C (18.4 mN/m -> N/m)
T = 272 + 273.15    # T: Super heated temperature of the liquid (272 °C -> K)
P_l = 101325        # P_l: Ambient pressure of the bulk liquid (1 atm -> Pa)
T0 = 100 + 273.15   # T0: Normal boiling point of water at 1 atm (100 °C -> K)
R = 8.314           # R: Ideal gas constant in J/(mol·K)

# === Function to Calculate Bubble Radius ===
# This function calculates the radius of a vapor bubble using the rearranged Thompson Equation
def calculate_bubble_radius(H_vap, T0, T, R, gamma, P_l):
    """Calculates the equilibrium radius of a vapor bubble using the Thompson equation."""
    r = (2 * gamma) / (P_l * (np.exp(H_vap * (1/T0 - 1/T) / R) - 1))
    return r

# Calculate the radius of the vapor bubble
radius = calculate_bubble_radius(H_vap, T0, T, R, gamma, P_l)

# Convert radius to nanometers for better readability
radius_nm = radius * 1e9  # Convert from meters to nanometers

# === Display the Result ===
print("\033[1m")  # Bold text
print(f'Radius of the vapor bubble: {radius_nm:.2f} nm') # Display radius in nanometers with 2 decimal places
print("\033[0m")  # Reset text formatting
```

## **3. NaCl Particles Super Saturation** ##  
---
NaCl particles (density = 2.17 g/$cm^3$) with a specific surface area of 4.25 x $10^5$ $cm^2$/g show a supersaturation of 6.71 % in ethanol (T=25 $\degree C$).  

**a)** Assuming the crystals are uniform spheres, find their radius (r)  
**b)** Estimate the NaCl / ethanol interfacial tension ($\gamma$) from this solubility data.

### **3a. Calculation of Particle Radius**  ###  
**Theoretical Framework:** The specific surface area ($A_{sp}$) of uniform spheres is geometrically related to their density ($\rho$) and radius ($R_s$).  
  
The total surface area of a single spere is $4\pi R_s^2$. The mass of a single sphere is $\text{Volume x Density} = \frac{4}{3}\pi R_s^3 \rho$.  
  
$A_{sp}$ is defined as the total surface area per unit mass:

$$
A_{sp} = \frac{\text{Surface Area}}{\text{Mass}} = \frac{4 \pi R_s^2}{\frac{4}{3}\pi R_s^3 \rho} = \frac{3}{\rho R_s}
$$  

**Equation Rearrangement:** We can rearrange this formula to solve for the radius, $R_s$:  
  
$$
R_s = \frac{3}{A_{sp}\rho}
$$


```python
# === Define Constants and Given Data for NaCl Particles ===
A_sp = 4.25e5       # A_sp: Specific surface area of NaCl particles in cm^2/g
rho_NaCl = 2.17     # rho_NaCl: Density of NaCl in g/cm^3

# === Calculations ===
R_s_cm = 3.0 / (A_sp * rho_NaCl)   # Radius Rs is calculated in cm from the rearranged formula

# Convert radius from cm to nm
R_s_nm = R_s_cm * 1e7              # Convert to nm

# === Display the Result ===
print("\033[1m")  # Bold text
print(f'Radius of NaCl particles: {R_s_nm:.2f} nm')
print("\033[0m")  # Reset text formatting
```

### **3b. Estimation of Interfacial Tension** ###  
  
**Theoretical Framework:** The increased solubility of small particles compared to a flat surface is described by the **Kelvin Equation**. This equation relates interfacial tension ($\gamma$) to the particle radius($R_s$) and the degree of supersaturation ($S / S_0$).  
  
For an electrolyte that dissociates into m cations and n anions (like NaCl, where m=1, n=1), the equation can be found on **page 263 Equation (42)**:  
  
$$
\frac{2M\gamma}{\rho R_s} = RT ln(\frac{a}{a_0}) = (m + n) RTln(\frac{S}{S_0})
$$

We will focus on the following:  

$$
\frac{2M\gamma}{\rho R_s} = (m + n) RTln(\frac{S}{S_0})
$$  

**Equation Rearrangement:** We rearrange the equation to solve for the interfacial tension, $\gamma$:  

$$
\gamma = \frac{(m+n)RT\rho R_s ln(\frac{S}{S_0})}{2M}
$$


```python
# === Define Constants and Given Data ===

# For NaCl, which dissociates into Na+ and Cl-, m=1 and n=1
m = 1.0                         # number of cations for NaCl
n = 1.0                         # number of anions for NaCl

# Supersaturation is given as 6.71%
supersaturation = 0.0671        # Supersaturation value
S_S0 = 1.0 + supersaturation    # The ratio of solubility (S) to bulk solubility (S0)

# Other given values
R = 8.314                       # R: Ideal Gas constant in J/(mol·K)
T = 25 + 273.15                 # T: Temperature in Kelvin (25 °C -> K)
rho = 2170                      # rho: Density of NaCl (2.17 g/cm^3 -> 2170 kg/m^3)
R_s = R_s_nm * 1e-9             # R_s: radius of particles calculated in part (a), convert from nm to m.
M = 58.44e-3                    # M: Molar mass of NaCl (58.44 g/mol -> kg/mol)

# === Calculation ===
gamma_Nm = ((m + n) * R * T * rho * R_s * np.log(S_S0)) / (2 * M) # Interfacial tension (gamma) is calculated

gamma_mNm = gamma_Nm * 1e3  # Convert from N/m to mN/m

# === Display the Result ===
print("\033[1m")  # Bold text
print(f'NaCl / Ethanol interfacial tension: {gamma_mNm:.2f} mN/m') 
print("\033[0m")  # Reset text formatting
```

## **4. Isoamyl Alcohol Water Interaction** ##  
---  
**Question:** The surface tension of isoamyl alcohol (liquid "B") and its interfacial tension with water (liquid "A") at T = 25 $\degree C$ are listed below. $\gamma_B$ is the surface tension of pure B, while $\gamma_{B(A)}$ represents the surface tension of A that is saturated with B, and $\gamma_{AB}$ represents the A/B interfacial tension. The surface tension of water at 25 $\degree C$ is 72.2 nM/m.  
  
**Data:**
* $\gamma_A$ (pure water) = 72.2 mN/m
* $\gamma_B$ (pure isoamyl alcohol) = 23.7 mN/m
* $\gamma_{A(B)}$ (water saturated with B) = 25.9 mN/m
* $\gamma_{B(A)}$ (isoamyl alcohol saturated with A) = 23.6 mN/m
* $\gamma_{AB}$ (interfacial) = 5.0 mN/m

**a)** Will isoamyl alcohol spread on water?

**b)** Will water spread on isoamyl alcohol?

**c)** Will either situation produce the “strange” behavior where a liquid initially spreads, only to retract into a lens after time?


### **4a) Will isonamyl alcohol spread on water?** ###  
   
**Theoretical Framework:** Spontaneous spreading occurs if the process leads to a decrease in the total surface free energy. This condition is quantified by the **spreading coefficient (S)**. A positive value (S > 0) indicates spontaneous spreading.  
  
For a liquid B (isoamyl alcohol) spreading on a substrate A (water), the equation is:  

$$
S_{B/A} = \gamma_A - (\gamma_B + \gamma_{AB})
$$  
  
This can be interpreted as the surface tension of the substrate minus the sum of the surface tensions of the spreading liquid and the new interface that is created.


```python
# === Define Given Data ===
gamma_water = 72.2          # Surface tension of water (substrate) in mN/m
gamma_isoamyl = 23.7        # Surface tension of isonamyl alcohol (spreading liquid) in mN/m
gamma_interfacial = 5.0     # Interfacial tension between A and B in mN/m

# === Calculation ===
S_BA = gamma_water - gamma_isoamyl - gamma_interfacial # Calculate spreading coefficient for B on A


# === Display the Result and Conclusion ===
print("\033[1m")  # Bold text
print(f'Spreading Coefficient (S_BA): {S_BA:.2f} mN/m')
print("\033[0m")  # Reset text formatting

if S_BA > 0:
    print("\033[1m")  # Bold text
    print("The isonamyl alcohol will spread spontaneously on the water surface.")
    print("\033[0m")  # Reset text formatting
else:
    print("\033[1m")  # Bold text
    print("The isonamyl alcohol will not spread spontaneously on the water surface.")
    print("\033[0m")  # Reset text formatting
```

### **4b) Will water spread on isoamyl alcohol?** ###  
  
**Theoretical Framework:** The same principle applies, but the roles of substrate and spreading liquid are reversed. For a liquid A (water) spreading on a substrate B (isoamyl alcohol), the equation is: 

$$ 
S_{A/B} = \gamma_B - (\gamma_A + \gamma_{AB})
$$

```python
# === Define Given Data ===
# Same data as before, but roles of substrate and spreading liquid are reversed

# === Calculation ===
S_AB = gamma_isoamyl - gamma_water - gamma_interfacial # Calculate spreading coefficient for A on B

# === Display the Result and Conclusion ===
print("\033[1m")  # Bold text
print(f'Spreading Coefficient (S_AB): {S_AB:.2f} mN/m')
print("\033[0m")  # Reset text formatting

if S_AB > 0:
    print("\033[1m")  # Bold text
    print("The isonamyl alcohol will spread spontaneously on the water surface.")
    print("\033[0m")  # Reset text formatting
else:
    print("\033[1m")  # Bold text
    print("The isonamyl alcohol will not spread spontaneously on the water surface.")
    print("\033[0m")  # Reset text formatting
```

### **4c) Will the retraction behavior occur?** ###  
**Theoretical Framework:** The "strange" behavior of initial spreading followed by rtraction into a lens ocurs if the **initial spreading coefficient is positive**, but the **final spreading coefficient (after the liquids have mutually saturated each other) is negative**.  
  
From part (a), we know isoamyl alcohol spread on water ($S_{B/A} > 0$). We must now calculate the final spreading coefficient, $S_{B/A'}'$ using the surface tension values for the saturated phases.  
  
The final spreading coefficient for B on A is:   
  
$$
S_{B/A}' = \gamma_{A(B)} - (\gamma_{B(A)} + \gamma_{AB})
$$

```python
# === Define Given Data ===
gamma_water_saturated = 25.9    # Surface tension of water saturated with isonamyl alcohol in mN/m
gamma_isoamyl_saturated = 23.6  # Surface tension of isonamyl alcohol saturated with water in mN/m
gamma_interfacial = 5.0         # Interfacial tension between phases in mN/m

# === Calculation ===
S_AB = gamma_water_saturated - gamma_isoamyl_saturated - gamma_interfacial # Calculate the final spreading coefficient for B on A using saturated values.

# === Display the Result and Conclusion ===
print("\033[1m")  # Bold text
print(f'Spreading Coefficient (S_AB): {S_AB:.2f} mN/m')
print("\033[0m")  # Reset text formatting

if S_AB < 0:
    print("\033[1m")  # Bold text
    print("Yes, the spreading of isoamyl alcohol on water will produce the strange behavior observed.")
    print("\nReasoning:")
    print("1. Initial State (S > 0): Isoamyl alcohol initiall spread across the water surface to lower the system's energy.")
    print("2. Mutual Saturation: Over time, the two liquids mutually saturate each other, changing their surface tensions.")
    print("3. Final State (S < 0): After saturation, the final spreading coefficient becomes negative, indicating that the system's energy is now lower when the isoamyl alcohol retracts into a lens.")
    print("\033[0m")  # Reset text formatting
else:
    print("\033[1m")  # Bold text
    print("No, the spreading of isoamyl alcohol on water will not produce the strange behavior observed.")
    print("\033[0m")  # Reset text formatting
```

### **5a) Estimating Critical Surface Tension with a Zisman Plot** ###  
  
**Theoretical Framework:** The **Zisman plot** is an emperical method to estimate the **critical surface tension ($\gamma_c$)** of a low-energy solif. The critical surface tension is defined as the surface tension of a hypothetical liquid that would completely wet the solid (have a contact angle of $\theta = 0\degree$).  
  
The method invovles plotting $cos\theta$ for a series of non-interacting liquids against their own surface tensions, $\gamma_L$. A linear trend is typically observed, which can be extrapolated to the point where $cos\theta = 1$ (since $cos(0\degree) = 1$). The corresponding $\gamma_L$ value at this intercept is the critical surface tensions, $\gamma_c$.  
  
**Methodology:**  
  
1. Plot $\theta$ (y-axis) versus $\gamma_L$ (x-axis).
2. Perform a linear regression on the data points to get an equation of the form $cos\theta = 0 m*\gamma_L + c$.  
3. Extrapolate this line to find the value of $\gamma_L$ when $cos\theta = 1$. This value is $\gamma_c$.  
  
**Note:** 
The data for **pentane** was ignored while drawing the regression line and calculating the $R^2$ value.  
This was done to avoid biasing the fit toward the critical point as pentane has a contact angle of 0.

```python
# === Linear Regression and Extapolation for Zisman Plot ===
def zisman_plot(surface_tensions, cos_thetas):
    """
    Generates a Zisman plot with surface tension on the x-axis 
    and cos(theta) on the y-axis.
    """
    # --- Analysis ---
    # Fit a linear model: cos(theta) = f(surface_tension) without including the value where cos(theta)=1
    # This is to avoid biasing the fit towards the critical point.
    coeffs_excluding_1 = np.polyfit(surface_tensions[:-1], cos_thetas[:-1], 1)
    fit_line_excluding_1 = np.poly1d(coeffs_excluding_1)

    # Calculate critical surface tension by finding the x-intercept where cos(theta) = 1.
    # The equation is y = mx + c, so we find x when y=1.
    # 1 = m*x + c  => x = (1 - c) / m
    m, c = coeffs_excluding_1
    critical_surface_tension = (1 - c) / m

    # Generate a range for the fit line that extends to the critical point
    # to ensure the extrapolation is visible on the plot.
    extended_surface_tensions = np.append(surface_tensions, critical_surface_tension)
    extended_surface_tensions.sort()
    
    # --- Plotting ---
    plt.figure(figsize=(8, 6))
    
    # Plot the raw data points
    plt.scatter(surface_tensions, cos_thetas, color='blue', label='Data Points')
    
    # Plot the extrapolated fit line
    plt.plot(extended_surface_tensions, fit_line_excluding_1(extended_surface_tensions), color='red', label='Fit Line (Extrapolated)')
    
    # Add a horizontal line at cos(theta) = 1 for reference
    plt.axhline(y=1, color='gray', linestyle=':', label='cos(θ) = 1')
    
    # Add a vertical line to mark the critical surface tension
    plt.axvline(x=critical_surface_tension, color='green', linestyle='--', 
                label=f'Critical Surface Tension: {critical_surface_tension:.2f} mN/m')
    
    # Add titles and labels
    plt.title('Zisman Plot')
    plt.xlabel('Surface Tension (mN/m)')
    plt.ylabel('cos(θ)')

    # Add r^2 value to the plot
    correlation_matrix = np.corrcoef(surface_tensions[:-1], cos_thetas[:-1])
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2
    plt.text(min(extended_surface_tensions), 0.5, f'$R^2 = {r_squared:.3f}$', fontsize=12, color='black')
    plt.legend()
    plt.grid(True)
    
    # Display the plot
    plt.show()

    return critical_surface_tension

# === Data preparation ===
# surface_tensions: Surface tensions of the homologous series of liquids (in mN/m)
surface_tensions = np.array([27.6, 26.7, 25.4, 24.7, 23.9, 22.9, 21.8, 20.3, 18.4, 16.0])

# thetas_deg: Corresponding contact angles (in degrees)
thetas_deg = np.array([46, 44, 42, 39, 35, 32, 26, 21, 12, 0])

# Calculate the cosine of the angles (NumPy's cos function requires radians)
cos_thetas = np.cos(np.radians(thetas_deg))

# --- Execution ---
# Generate the plot and get the critical surface tension
critical_surface_tension = zisman_plot(surface_tensions, cos_thetas)

# === Print the final result ===
print("\033[1m")  # Bold text
print(f'Calculated Critical Surface Tension: {critical_surface_tension:.2f} mN/m')
print("\033[0m")  # Unbold text
```
