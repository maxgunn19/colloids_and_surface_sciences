## **1. Two Spherical Bubbles** ##  
Two sperical bubbles come into contact and fuse in a liquid of hydrostatic pressure P. The radius of the larger bubble is $R_a$ and that of the smaller bubble is $R_b$. There is a curved septum where the two bubbles are in contact. The radius of curvature of the septum is $R_s$. Derive an equation that relates $R_s$ to $R_a$ and $R_b$. (Note: the septum is a thin liquid film, so it has a vapor/liquid interface on both sides.) Why is the septum shaped the way that it is?

![Bubble Image](./HW_2_1_image.png)


**a)** Derive an equation that relates $R_s$ to $R_a$ and $R_b$.

**1a) EXPLANATION**  
  
The Young-Laplace Equation is the following:

$$
\Delta P = \gamma(\frac{1}{R_1} + \frac{1}{R_2})
$$

For a perfect sphere, both radii are equal to one another ($R_1 = R_2 = R$). This makes the Laplace Pressure for the sphere the following:

$$
\Delta P = \gamma (\frac{1}{R} + \frac{1}{R}) = \frac{2 \gamma}{R}
$$

For the **larger bubble (a)**, the internal pressure $P_a$ is:

$$
P_a - P = \frac{2 \gamma}{R_a} \implies P_a = P + \frac{2 \gamma}{R_a} \tag{1}
$$

For the **smaller bubble (b)**, the internal pressure $P_b$ is:

$$
P_b - P = \frac{2 \gamma}{R_b} \implies P_b = P + \frac{2 \gamma}{R_b} \tag{2}
$$

The septum is the film that separates the two bubbles. This film has a liquid/vapor interface on both sides. Therefore, the pressure difference it can sustain is twice that of a single interface:

$$
\Delta P_{septum} = \frac{4 \gamma}{R_s} \tag{3}
$$

The pressure difference must be equal to the difference between the internal pressures of the two bubbles. Since $R_b < R_a$, the pressure $P_b$ will be greater than $P_a$. So, the pressure difference across the septum is the following:

$$
P_b - P_a = \frac{4 \gamma}{R_s}
$$

We can now find another expression for the pressure difference $P_b - P_a$ using **Equation 1** and **Equation 2**:

$$
P_b - P_a = (P + \frac{2 \gamma}{R_b}) - (P + \frac{2 \gamma}{R_a})
$$

Simplifying we get:

$$
P_b - P_a = \frac{2\gamma}{R_b}-\frac{2\gamma}{R_a}
$$

We can now plug **Equation 3** in:

$$
\frac{4\gamma}{R_s} = \frac{2\gamma}{R_b} - \frac{2\gamma}{R_a}
$$

Dividing by $2\gamma$ we get:

$$
\frac{2}{R_s} = \frac{1}{R_b} - \frac{1}{R_a}
$$

We can now solve for **R_s** in relation to **R_a** and **R_b**.

$$
R_s = \frac{2R_aR_b}{R_a - R_b}
$$

**b)** Why is the septum shaped the way it is?

**2b) EXPLANATION**  
The shape of the septum is due to the pressure difference between the two bubbles.  
  
* **Pressure and Radius:** The Laplace Equation shows that the pressure inside a bubble is **inversly proportional** to its radius ($\Delta P \propto 1/R$)
* **Higher Pressure in Smaller Bubble:** Because the smaller bubble has a smaller radius ($R_b < R_a$), it has a **higher internal pressure** ($P_b > P_a).
* **Resulting Force:** When the bubbles touch, the pressure difference creates a net force on the wall between them. The higher pressure of the smaller bubble pushes the septum into the larger bubble.  
  
The septum is curved **bulging into the larger bubble**, with its center of curvature located inside the smaller bubble. The exact radius of the curvature, $R_s$, is determined by the balance between the surface tension forces and the pressure difference.

## **2. Vapor Bubble Radius** ##  
Show that the **Thompson Equation** for nucleation of vapor bubbles in a boiling liquid is:  
$$
\Delta H_{vap} [\frac{1}{T_0}-\frac{1}{T}] = R \ln[\frac{\frac{2\gamma}{r} + P_l}{P_l}]
$$  
where $P_l$ is the pressure in the bulk liquid phase, set by the ambient pressure of the surrounding gas. If the heat of vaporization is **31 kJ/mol** and the surface tension of the water is **18.4 mN/m** at **272 $\degree C$**, find the radius of the vapor bubble in equilibrium with liquid at an ambient pressure of **1 atm** and **T = 272 $\degree C$**.

**a)** Show that the **Thompson Equation** for nucleation of vapor bubbles in a boiling liquid is:  
$$
\Delta H_{vap} [\frac{1}{T_0}-\frac{1}{T}] = R \ln[\frac{\frac{2\gamma}{r} + P_l}{P_l}]
$$  

**2a) EXPLANATION**  
  
We will first start with the **Gibbs Duhem Equation** for both the liquid (l) and the gas (g)
$$
\bar{S_l} dT - \bar{V_l} dP_l + d\mu_l = 0
$$

$$
\bar{S_g} dT - \bar{V_g} dP_g + d\mu_g = 0
$$

They both equal 0 so they can be set equal to one another:

$$
\bar{S_l} dT - \bar{V_l} dP_l + d\mu_l = \bar{S_g} dT - \bar{V_g} dP_g + d\mu_g
$$

Now take the difference:

$$
[\bar{S_l} - \bar{S_g}]dT - \bar{V_l}dP_l + \bar{V_g}dP_g + d[\mu_l - \mu_g] = 0
$$

At equilibrium, the chemical potential difference is zero ($[d\mu_l - \mu_g] = 0$). That leaves us with the following: 

$$
[\bar{S_l} - \bar{S_g}]dT - \bar{V_l}dP_l + \bar{V_g}dP_g  = 0 
$$
  
After simplifying we have the following:

$$
[\bar{S_l} - \bar{S_g}]dT = \bar{V_l}dP_l - \bar{V_g}dP_g \tag{1}
$$

**Clausius Equation**: Relates entropy change of vaporization to the enthalpy of vaporization ($\Delta H_{vap}$).

$$
\bar{S_g} - \bar{S_l} = \frac{\Delta H_{vap}}{T} \tag{2}
$$

**Laplace Pressure Equation**: Describes the pressure difference across the curved surface of the bubble due to surface tension ($\gamma$).

$$
P_g - P_l = \frac{2\delta}{r} \implies P_g = P_l + \frac{2\gamma}{r} \tag{3}

$$

**Ideal Gas Law**: We assume the vapor in the bubble behaves as an ideal gas.
# but why #

$$
\bar{V}_g = \frac{RT}{P_g} \tag{4}
$$

We can assume the change external liquid pressure $dP_l$ is << $dP_g$ so $dP_l \approx 0$. After applying this assumption to **Equation 1** we get:

$$
[\bar{S_l} - \bar{S_g}]dT = - \bar{V_g}dP_g 
$$

We can now substitute in the **Clausius Equation (2)**:

$$
(-\frac{\Delta H_{vap}}{T})dT = - \bar{V}_g dP_g
$$

Simplifying we get:

$$
(\frac{\Delta H_{vap}}{T})dT = \bar{V}_g dP_g
$$

We can assume the vapor in the bubble behaves as an idea gas. That means that we can substitute in the **Ideal Gas Law (4)**:

$$
\frac{\Delta H_{vap}}{T} dT = (\frac{RT}{P_g})dP_g
$$

Now we can separate variables to prepare for integration. We will rearrange and get all the temperature (T) terms on one side and all the pressure ($P_g$) terms on the other. We find the following:

$$
\frac{\Delta H_{vap}}{R}\frac{dT}{T^2} = \frac{dP_g}{P_g}
$$

We will integrate from the normal boiling point ($T_0$) to the final superheated temperature (T) and from the normal boiling point (which is the external liquid pressure, $P_l$) to the pressure inside the bubble ($P_g$).

$$
\frac{\Delta H_{vap}}{R} \int_{T_0}^{T} \frac{dT}{T^2} = \int_{P_l}^{P_g} \frac{dP}{P}
$$

After performing the integration we can evaluate:

$$
\frac{\Delta H_{vap}}{R}[-\frac{1}{T}]_{T_0}^{T} = [\ln(P)]_{P_l}^{P_g}
$$

After applying the limits we find:

$$
\frac{\Delta H_{vap}}{R}(-\frac{1}{T}-(-\frac{1}{T_0})) = \ln (P_g) - \ln (P_l)
$$

Simplifying both sides gives us the **Clausius-Clapeyron Equation**:  

$$
\frac{\Delta H_{vap}}{R}(\frac{1}{T_0} - \frac{1}{T}) = \ln(\frac{P_g}{P_l})
$$

We can now substitute in the **Laplace Pressure Equation (3)**:

$$
\frac{\Delta H_{vap}}{R}(\frac{1}{T_0} - \frac{1}{T}) = \ln(\frac{P_l + \frac{2\gamma}{r}}{P_l})
$$

Moving the R over to the right side we are now left with the **Thompson Equation for nucleation of a vapor bubble**:

$$
\Delta H_{vap} [\frac{1}{T_0}-\frac{1}{T}] = R \ln[\frac{\frac{2\gamma}{r} + P_l}{P_l}]
$$  


**b)** If the heat of vaporization is **31 kJ/mol** and the surface tension of the water is **18.4 mN/m** at **272** $\degree C$, find the **radius of the vapor bubble** in equilibrium with liquid at an ambient pressure of **1 atm** and **T = 272** $\degree C$.

```python
# Constants and Data
H_vap = 31e3        # Heat of vaporization in J/mol (31 kJ/mol)
gamma = 18.4e-3     # Surface tension in N/m (18.4 mN/m)
T = 272 + 273.15    # Temperature in K (272 °C)
P_l = 101325        # Ambient pressure in Pa (1 atm)
T0 = 100 + 273.15   # Boiling point of water in K (100 °C)
R = 8.314           # Gas constant in J/(mol·K)

# Rearranging the Thompson equation to solve for r
def calculate_bubble_radius(H_vap, T0, T, R, gamma, P_l):
    r = (2 * gamma) / (P_l * (np.exp(H_vap * (1/T0 - 1/T) / R) - 1))
    return r

# Calculate the radius of the vapor bubble
radius = calculate_bubble_radius(H_vap, T0, T, R, gamma, P_l)

print("\033[1m")  # Bold text
print(f'Radius of the vapor bubble: {radius*1e9:.2f} nm')  # Convert to nm for better readability
print("\033[0m")  # Reset text formatting
```

**2b) EXPLANATION**  

We will start with the **Thompson Equation for nucleation of a vapor bubble**:

$$
\Delta H_{vap} [\frac{1}{T_0}-\frac{1}{T}] = R \ln[\frac{\frac{2\gamma}{r} + P_l}{P_l}]
$$

**<center>We need to rearrange this equation to solve for r</center>**

We can first simplify the **natural log (ln)** portion of the equation:  

$$
\frac{\frac{2\gamma}{r} + P_l}{P_l} \implies \frac{2\gamma}{rP_l} + \frac{P_l}{P_l} \implies 1 + \frac{2\gamma}{rP_l}
$$

Next we can isolate the **natural log (ln)** and insert its simplified version:

$$
\frac{\Delta H_{vap}}{R}[\frac{1}{T_0} - \frac{1}{T}] = ln[1 + \frac{2\gamma}{rP_l}]
$$

Next we can eliminate the **natural log (ln)**:

$$
\exp(\frac{\Delta H_{vap}}{R}[\frac{1}{T_0}-\frac{1}{T}]) = 1 + \frac{2\gamma}{rP_l}
$$

Finally we can isolate and solve for r:

$$
r = \frac{2\gamma}{P_l(\exp(\frac{\Delta H_{vap}}{R}[\frac{1}{T_0}-\frac{1}{T}])-1)}
$$

## **3. NaCl Particles Super Saturation** ##  
NaCl particles (density = 2.17 g/$cm^3$) with a specific surface area of 4.25 x $10^5$ $cm^2$/g show a supersaturation of 6.71 % in ethanol (T=25 $\degree C$). Assuming the crystals are uniform spheres, find their radius (r) and estimate the NaCl / ethanol interfacial tension ($\gamma$) from this solubility data.

**a)** Assuming the crystals are uniform spheres, find their radius (r)

```python
A_sp = 4.25e5       # Specific surface area in cm^2/g
rho_NaCl = 2.17     # Density of NaCl in g/cm^3

R_s = 3.0 / (A_sp * rho_NaCl)   # Radius in cm
R_s_nm = R_s * 1e7              # Convert to nm

print("\033[1m")  # Bold text
print(f'Radius of NaCl particles: {R_s_nm:.2f} nm')
print("\033[0m")  # Reset text formatting
```

**3a. EXPLANATION**  
The textbook provides a direct relationship between the **specific surface area ($A_{sp}$)**, the **density ($\rho$)** of the particles, and their **radius ($R_s$)**.  
  
**Equation (2) on page 9**:

$$
A_{sp} = \frac{3}{\rho R_s}
$$

This formula is derived from the basic geometric properties of a sphere.  
  
Rearrange the formula to solve for the unknown radius, $R_s$:  

$$
R_s = \frac{3}{A_{sp}\rho}
$$

**b)** Assuming the crustals are uniform spheres, estimate the NaCl / ethanol interfacial tension ($\gamma$).

```python
m = 1.0                         # m value for NaCl
n = 1.0                         # n value for NaCl
supersaturation = 0.0671        # Supersaturation value
S_S0 = 1.0 + supersaturation    # S/S0 value
R = 8.314                       # Gas constant in J/(mol·K)
T = 25 + 273.15                 # Temperature in K (25 °C)
rho = 2170                      # Density of NaCl in kg/m^3
R_s = R_s_nm * 1e-9             # Convert radius to meters
M = 58.44e-3                    # Molar mass of NaCl in kg/mol

gamma = (m + n) * (R * T * rho * R_s * np.log(S_S0)) / (2 * M)

print("\033[1m")  # Bold text
print(f'NaCl / Ethanol interfacial tension: {gamma*1e3:.2f} mN/m')  # Convert to mN/m for better readability
print("\033[0m")  # Reset text formatting
```

**3b. EXPLANATION**

The increased solubility of the small particles is due to their surface curvature. This is seen in the **Kelvin Equation**.  

**Equation (42) on page 263**:  

$$
\frac{2M\gamma}{\rho R_s} = RT ln(\frac{a}{a_0}) = (m + n) RTln(\frac{S}{S_0})
$$

We will focus on the following:  

$$
\frac{2M\gamma}{\rho R_s} = (m + n) RTln(\frac{S}{S_0})
$$  

The equation can be rearranged to solve for $\gamma$:  

$$
\gamma = \frac{(m+n)RT\rho R_s ln(\frac{S}{S_0})}{2M}
$$
  
**Defining each variable**
* (m+n): For NaCl **m = 1 and n = 1**.
* $S/S_0$: The "supersaturation of 6.71%" means that the solubility (S) is 1.0671 times the normal solubility $S_0$. So $\mathbf{S/S_0} = 1.0671$
* R: The ideal gas constant is **8.314 J/(mol*K)**.
* T: The absolute temperature is **25** $\degree C + 273.15$ = **298.15 K**.
* $\rho$: The density of NaCl is **2.17** $\mathbf{g/cm^3=2170kg/m^3}$.
* $R_s$: The radius calculated in the first part is 32.5nm = $\mathbf{3.25*10^-8}$.
* M: The molar mass is **58.44 g/mol**

a) Will isonamyl alcohol spread on water?  


```python
gamma_water = 72.2          # Surface tension of water in mN/m
gamma_isoamyl = 23.7        # Surface tension of isonamyl alcohol in mN/m
gamma_interfacial = 5.0     # Interfacial tension in mN/m

S_BA = gamma_water - gamma_isoamyl - gamma_interfacial

print("\033[1m")  # Bold text
print(f'Spreading Coefficient (S_BA): {S_BA:.2f} mN/m')
print("\033[0m")  # Reset text formatting

if S_BA > 0:
    print("\033[1m")  # Bold text
    print("The isonamyl alcohol will spread spontaneously on the water surface.")
    print("\033[0m")  # Reset text formatting
else:
    print("\033[1m")  # Bold text
    print("The isonamyl alcohol will not spread spontaneously on the water surface.")
    print("\033[0m")  # Reset text formatting
```

**4a) EXPLANATION**  
  
The **spreading coefficient (S)** is all that matters.
  
$$
-S = \gamma_{substrate} - \gamma_{spreading} - \gamma_{interfacial}
$$

* **S** is the spreading coefficient.  
* $\gamma_{substrate}$ is the surface tension of the liquid being spread upon (**water**).
* $\gamma_{spreading}$ is the surface tension of the liquid that is spreading (**isoamyl B**).
* $\gamma_{interfacial}$ is the surface tension between the two liquids (**interfacial AB**).

Spreading is spontaneous only when **S > 0**.

## **5. Critical Surface Tension of Teflon** ##

**a)** Estimate the critical surface tension for a Teflon surface with the following wetting characteristics. 

```python
# Zisman Plot
def zisman_plot(surface_tensions, cos_thetas):
    """
    Generates a Zisman plot with surface tension on the x-axis 
    and cos(theta) on the y-axis.
    """
    # Fit a linear model: cos(theta) = f(surface_tension)
    coeffs = np.polyfit(surface_tensions, cos_thetas, 1)
    fit_line = np.poly1d(coeffs)

    # Calculate critical surface tension by finding the x-intercept where cos(theta) = 1.
    # The equation is y = mx + c, so we find x when y=1.
    # 1 = m*x + c  => x = (1 - c) / m
    m, c = coeffs
    critical_surface_tension = (1 - c) / m

    # Generate a range for the fit line that extends to the critical point
    # to ensure the extrapolation is visible on the plot.
    extended_surface_tensions = np.append(surface_tensions, critical_surface_tension)
    extended_surface_tensions.sort()
    
    # --- Plotting ---
    plt.figure(figsize=(8, 6))
    
    # Plot the raw data points
    plt.scatter(surface_tensions, cos_thetas, color='blue', label='Data Points')
    
    # Plot the extrapolated fit line
    plt.plot(extended_surface_tensions, fit_line(extended_surface_tensions), color='red', label='Fit Line (Extrapolated)')
    
    # Add a horizontal line at cos(theta) = 1 for reference
    plt.axhline(y=1, color='gray', linestyle=':', label='cos(θ) = 1')
    
    # Add a vertical line to mark the critical surface tension
    plt.axvline(x=critical_surface_tension, color='green', linestyle='--', 
                label=f'Critical Surface Tension: {critical_surface_tension:.2f} mN/m')
    
    # Add titles and labels
    plt.title('Zisman Plot')
    plt.xlabel('Surface Tension (mN/m)')
    plt.ylabel('cos(θ)')

    # Add r^2 value to the plot
    correlation_matrix = np.corrcoef(surface_tensions, cos_thetas)
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2
    plt.text(min(extended_surface_tensions), 0.5, f'$R^2 = {r_squared:.3f}$', fontsize=12, color='black')
    plt.legend()
    plt.grid(True)
    
    # Display the plot
    plt.show()

    return critical_surface_tension

# --- Data ---
# Surface tensions of the liquids (in mN/m)
surface_tensions = np.array([27.6, 26.7, 25.4, 24.7, 23.9, 22.9, 21.8, 20.3, 18.4, 16.0])
# Corresponding contact angles (in degrees)
thetas_deg = np.array([46, 44, 42, 39, 35, 32, 26, 21, 12, 0])

# Calculate the cosine of the angles (NumPy's cos function requires radians)
cos_thetas = np.cos(np.radians(thetas_deg))

# --- Execution ---
# Generate the plot and get the critical surface tension
critical_surface_tension = zisman_plot(surface_tensions, cos_thetas)

# Print the final result

print("\033[1m")  # Bold text
print(f'Calculated Critical Surface Tension: {critical_surface_tension:.2f} mN/m')
print("\033[0m")  # Unbold text
```

**5a) EXPLANATION**  
The **Zisman plot** is a graphical method used to determine the **critical surface tension** ($\gamma_{LV critical}$) of a solid. Measuring $\gamma_{LV}$ is easy but measuring $\gamma_{SV}$ is hard. Using *one* kind of solid but *many* different kinds of liquid allows one to find where $\gamma_{SV}=\gamma_{LV critical}$ which where the contact angle would be $0\degree$, which corresponds to $\cos(\theta)=1$

**b)** Is this value a good estimate of the surface energy (tension) of the Teflon?

**YES**, in literature other people have found the value for teflon to be between 18-20 mN/m. The Zisman plot is considered reliable for low-energy surfaces. The plot has an $R^2$ value of **0.965** which is considered to be high and a good indicator of the fit line being accurate which in turn indicates a more accurate calculation of the critical surface tension.
